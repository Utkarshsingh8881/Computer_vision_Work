```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


```

```python
hist1 = np.array([[0.1,0.2,0.3],
                 [0.2,0.3,0.1],
                 [0.1,0.1,0.0]], dtype = np.float32)
hist2 = np.array([[0.3,0.2,0.1],
                 [0.2,0.2,0.1],
                 [0.0,0.1,0.1]], dtype = np.float32)
```

```python
hist1 /= hist1.sum()
hist2 /= hist2.sum()
```

```python
emd_distance = cv2.EMD(hist1,hist2,cv2.DIST_L2)
```

```python
print("EARTH Mover's Distance:" ,emd_distance)
plt.imshow(hist1,cmap='Blues',alpha=0.5, extent = (0, 2, 0, 2),origin = 'upper')
plt.imshow(hist2,cmap='Reds',alpha=0.5, extent = (0, 2, 0, 2),origin = 'upper')
plt.xlabel('Bins (x)')
plt.ylabel('Bins (y)')
plt.colorbar(label ='probability')
plt.show()
```
